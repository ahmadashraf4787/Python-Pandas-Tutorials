# Some Examples Using Loc

```python
df.head()
```

```python
df.loc[0, :]
```

```python
df.loc[0:2, :]
```

```python
df.loc[[0,3,5], :]
```

```python
df.loc[:, 'price']
```

# Some Examples Using iLoc

```python
df.head()
```

```python
df.iloc[0:3, :]
```

```python
df.iloc[0:3, 0:5]
```

```python
df.iloc[[0,3,5], [0, 3, 5]]
```

```python
df.loc[[0,1,2,3], :]
```

```python
df.loc[0:3, :]
```

```python
df.loc[[1,3,5], 'price': 'mainroad']
```

```python
df.loc[[2,4,6], ['guestroom','hotwaterheating','parking']]
```

```python
df.loc[:, ['guestroom','hotwaterheating','parking']]
```

```python
df.head()
```

```python
df.loc[df['mainroad'] == 'yes', 'price']
```

```python
df[df['mainroad'] == 'yes']['price']
```
