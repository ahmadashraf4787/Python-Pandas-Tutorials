#  How to perform Data Frame Slicing in python pandas?

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Housing (1).csv")
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

# Extracting  the column as a Series and  as a Data frame

```python
df['airconditioning']
```

```python
type(df['airconditioning'])
```

```python
df[['airconditioning']]
```

```python
type(df[['airconditioning']])
```

# Slicing data frame using a few columns names

```python
df.head()
```

```python
df[['area','bedrooms','bathrooms']]
```

# Slicing data frame using a few rows

```python
df[0:5]
```

# Slicing data frame using both rows, and columns

```python
df[['area','bedrooms','bathrooms']][0:5]
```

# Slicing data frame using with a condion

```python
df[df['area']>8000]
```

```python
df[df['mainroad'] == 'yes']
```

# Slicing data frame using with multiple condions using AND operation

```python
df[(df['area']> 10000) & (df['mainroad'] == 'yes')]
```

#  Filtering Query function (for .query())

```python
df.query("area > 10000 and mainroad == 'yes'")
```

# Slicing the data frame using .values function

```python
df.values[0:5]
```

# Slicing the  data frame using .columns list

```python
df[df.columns[0:5]]
```

# Slicing data frame using .head(), tail() and .sample() commands

```python
df.head(2)
```

```python
df.tail(2)
```

```python
df.sample(2)
```

# Filtering  using select_dtypes command:

```python
df.select_dtypes(include = 'int')
```

```python
df.select_dtypes(include = 'object')
```
